```python
outliers=[]
c1marks=df['COURSE 1 MARKS']
def detect_outliers_iqr(data):
    data=sorted(data)
    q1=np.percentile(data, 25)
    q2=np.percentile(data, 75)
    IQR=q2-q1
    lower_bound=q1-(1.5*IQR)
    upper_bound=q2-(1.5*IQR)
    for i in data:
        if(i < lower_bound or upper_bound > i):
          outliers.append(i)
    return outliers
c1marks_outliers = detect_outliers_iqr(c1marks)
print("outliers from IQR method are :", c1marks_outliers)
```
